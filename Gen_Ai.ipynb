```python
# Install dependencies
!pip install -q transformers accelerate gradio

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
import gradio as gr

# Load model and tokenizer
model_id = "HuggingFaceH4/zephyr-7b-alpha"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype=torch.bfloat16, device_map="auto")

# Create text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Function to generate assistant response
def generate_reply(history, user_input):
    # Build the prompt from chat history
    chat_prompt = "<|system|>\nYou are a helpful assistant.\n"
    for msg in history:
        chat_prompt += f"<|user|>\n{msg[0]}\n<|assistant|>\n{msg[1]}\n"
    chat_prompt += f"<|user|>\n{user_input}\n<|assistant|>\n"

    # Generate the response
    output = generator(chat_prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_p=0.9)[0]["generated_text"]

    # Extract only the assistant reply
    reply = output.split("<|assistant|>\n")[-1].strip()
    return reply

# Gradio chatbot interface
with gr.Blocks() as demo:
    gr.Markdown("## 🤖 ChatGPT-Style Text Generator using Zephyr-7B")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Your Message")
    clear = gr.Button("Clear Chat")

    history = gr.State([])

    def respond(user_message, history):
        reply = generate_reply(history, user_message)
        history.append((user_message, reply))
        return history, history

    msg.submit(respond, [msg, history], [chatbot, history])
    clear.click(lambda: ([], []), None, [chatbot, history])

demo.launch(share=True)
```

```python
# 🚀 All‑in‑one text‑to‑text demo (single cell)

# 1) Install (first run only)
!pip -q install --upgrade transformers==4.28.1 torch sentencepiece accelerate pytorch-lightning

# 2) Imports & model
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch, textwrap

MODEL_NAME = "t5-small"          # ▶ swap e.g. "google/flan-t5-base", "google/flan-t5-xl"
device     = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model     = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME).to(device)

# 3) Helper
def t2t(prompt, max_new_tokens=128, temperature=0.7, top_p=0.9):
    inputs  = tokenizer(prompt, return_tensors="pt").to(device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=temperature,
        top_p=top_p,
        no_repeat_ngram_size=3,
        early_stopping=True,
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

# 4) Quick demos
examples = [
    ("Translate EN→DE",
     "translate English to German: How are you feeling today?"),
    ("Summarize",
     "summarize: Deep learning models need large amounts of labeled data for accuracy, "
     "but collecting it is costly; data‑augmentation is a cheaper alternative."),
    ("Question Answering",
     "question: Who designed the Eiffel Tower?  "
     "context: The Eiffel Tower was designed and built by Gustave Eiffel's company."),
]

for title, prompt in examples:
    print(f"\n🟢 {title}\nPrompt: {prompt}\n🔵 Output: {t2t(prompt)}\n")

# 5) Interactive loop
print("Enter text‑to‑text prompts (type 'quit' to exit)\n")
try:
    while True:
        user_prompt = input(">>> ")
        if user_prompt.lower().strip() in {"quit", "exit"}:
            break
        print(textwrap.fill(t2t(user_prompt), width=100), end="\n\n")
except EOFError:
    pass
```

```python
!pip install transformers gradio torchaudio accelerate

import gradio as gr
import torch
from transformers import pipeline

# Load speech-to-text pipeline
pipe = pipeline("automatic-speech-recognition", model="openai/whisper-base")

# Function to transcribe audio
def transcribe(audio):
    text = pipe(audio)["text"]
    return text

# Gradio Interface
interface = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath", label="🎙 Speak now"),
    outputs=gr.Textbox(label="📝 Transcribed Text"),
    title="🗣 Speech-to-Text Converter (Free & Open Source)",
    description="Powered by OpenAI Whisper (via Hugging Face Transformers)"
)

interface.launch()
```

```python
!pip install diffusers==0.28.0 transformers==4.38.2 gradio accelerate safetensors huggingface_hub==0.20.2


import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

# Load the stable diffusion model from HuggingFace
pipe = StableDiffusionPipeline.from_pretrained(
"runwayml/stable-diffusion-v1-5",
torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Function to generate image
def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

# Gradio Interface
interface = gr.Interface(
fn=generate_image,
inputs=gr.Textbox(label="Enter your prompt"),
outputs="image",
title="🎨 Text-to-Image Generator (Free & Open Source)",
description="Powered by Stable Diffusion"
)

interface.launch()
```
